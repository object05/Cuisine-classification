```python
import pandas as pd                   #for data handling (espically with data_frames)
from sklearn import preprocessing     #for labeling cuisines (apply numerical labels or ids on cuisines )
from sklearn.feature_extraction.text import TfidfVectorizer #to convert text data to numericals 
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split  #for spilitting data set into test data set and train data set 
from sklearn.metrics import accuracy_score, confusion_matrix  #for confusion matrix and accuracy score
from sklearn.svm import SVC
from sklearn import svm #our training algorithm 
from sklearn.model_selection import GridSearchCV # optional(for hyper parameter tuning)
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
import os
import numpy as np
from Recipe2Cuisine_functions import preprocess, plot_confusion_matrix
import re
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')#in the future, labels will have to be numeric
```

### Import json train files and check training data

```python
#retrieving data from .json files
df_train = pd.read_json("./train.json")
df_test = pd.read_json("./test.json")
df_train.head()
```

### Also check testing data

```python
df_test.head()
```

## Some data stats, count by origin:

```python
# Counts of the number of cuisine labels 
fig, ax = plt.subplots()
fig.set_size_inches(18, 7)
df_train['cuisine'].value_counts().plot(kind='bar');
sns.set_style("whitegrid") 
plt.xticks(rotation = '90', fontsize = 17);
plt.yticks(fontsize = 17);
ax.set_xlabel('Cuisines', fontsize = 20, weight='bold');
ax.set_ylabel('Number of Labels', fontsize = 20, weight='bold');
```

### Process data, remove as much noise as possible, normalise text

```python
# Remove all symbols, extra spaces, leave only underscores and single spaces
for item in df_train['ingredients']:
    #item.replace("-", "_")
    for s in item:
        re.sub(r"-", "_", s)
        re.sub(r"[^a-z0-9_ ]", "", s)
        

        
for item in df_test['ingredients']:
    #item.replace("-", "_")
    for s in item:
        re.sub(r"-", "_", s)
        re.sub(r"[^a-z0-9_ ]", "", s)
```

### Converting data into right shape and format

```python
df_train['ingredients'] = df_train['ingredients'].apply(','.join) # this converts ingredients columns into arrays
df_test['ingredients'] = df_test['ingredients'].apply(','.join)   # for test data, above one is for train data
X_train = df_train['ingredients']  # assigning a new variable for ingredients array(for train data)
X_test = df_test['ingredients']    # assigning a new variable for ingredients array(for test data)
X_train.head()
```

```python
X_test.head()
```

### Converting data into something models will understand

```python
#all our parameters(i.e cuisine,ingredients) are in text format
#most of the machine learning algorithms cannot handle text data forms
#so we have to convert the text data into some numerical data without losing its quality 

encoder = preprocessing.LabelEncoder() #encoder for our y-values i.e cuisine
y_train_transformed = df_train['cuisine']##this labels each y_value(which is in text) to a number 
#y_train_transformed = encoder.fit_transform(df_train['cuisine'])

#as our X-values i.e ingredients are grouped as array for each cuisine 
#we need to vectorize or simply assign a set of numerical value to each element in array without losing its quality.
#one of the effective tool for this is TfidVectorizer

vec = TfidfVectorizer(binary = True).fit(X_train.values) #assign our vectoriser 
X_train_transformed = vec.transform(X_train.values) #applying vectorizer for x-values of train set 
X_test_transformed = vec.transform(X_test.values)   #applying vectorizer for x-values of test set 
```

### Splitting data into test and train (X and Y)

```python
#spiliting test-data set into further test-set and train-set
#In my view This spiliting process is for calculating accuracy_score 
X_for_train, X_for_test, y_for_train, y_for_test = train_test_split(X_train_transformed, y_train_transformed ,test_size= 0.25, random_state = 0)
```

### Prepare labels for confusion matrix

```python
# Classification results using confusion matrix
cuisines = df_train['cuisine'].value_counts().index
#labels = np.unique(y_for_test) #TODO use this for labels when other deprecated in confusion matrix
labels = [x for x in df_train['cuisine']]
labels = np.unique(labels)
print(labels)
```

# PREDICTIONS

## Random Forest Classification

```python
rfc = RandomForestClassifier(random_state=0,n_jobs=10)
rfc.fit(X_for_train, y_for_train)
y_pred = rfc.predict(X_for_test)  #predicting our test-set
accuracy = accuracy_score(y_for_test, y_pred)
print('accuracy_score = ', accuracy)
```

#### Confusion matrix

```python
RF_Confusion = confusion_matrix(y_for_test, rfc.predict(X_for_test), labels=labels) # construct confusion matrix 
plot_confusion_matrix(RF_Confusion, target_names=labels, title = 'Random Forest Confusion Matrix')
```

## eXtreme Gradient Boosting

```python
xgbc = XGBClassifier(max_depth=10, learning_rate=0.3,n_estimators=100,booster='gbtree',n_jobs=10,eval_metric='mlogloss',use_label_encoder=True)
xgbc.fit(X_for_train, y_for_train)
y_pred = xgbc.predict(X_for_test) 
accuracy = accuracy_score(y_for_test, y_pred)
print('accuracy_score = ', accuracy) #SEE A WARNING? go to 'labels' definition
```

```python
RF_Confusion = confusion_matrix(y_for_test, xgbc.predict(X_for_test), labels=labels) # construct confusion matrix 
plot_confusion_matrix(RF_Confusion, target_names=labels, title = 'eXtreme Gradient Boosting Confusion Matrix')
```

## Linear Support Vector Classification

```python
clf = svm.LinearSVC(C=0.5, max_iter=100, random_state=20, tol=0.5) #these are hyper parameters best suitable for this problem 
clf.fit(X_for_train, y_for_train) #training our train-set
y_pred = clf.predict(X_for_test)  #predicting our test-set
accuracy = accuracy_score(y_for_test, y_pred)
print('accuracy_score = ', accuracy)
```

```python
RF_Confusion = confusion_matrix(y_for_test, clf.predict(X_for_test), labels=labels) # construct confusion matrix 
plot_confusion_matrix(RF_Confusion, target_names=labels, title = 'Linear Support Vector Confusion Matrix')
```

# Can we find better parameters?

We can search for optimal parameters with hyper parameter tuning... we need to specify some parameters, then go through all combinations with a loop and save the results

### Random forest hyper parameter tuning

```python
param_grid = {'random_state': [0,50,100], 
                'n_estimators': [100,200,300],
                'max_depth': [None, 20, 30],
                'n_jobs':[10]}

# all the numbers in param_grid are for an optimal parameter tuning 
# tested with many possibilities and provided the best amongest them 
#note: these params vary from problem to problem. These are some of best suitable for this problem 
           
optimal_rfc = RandomForestClassifier()
param = optimal_rfc.get_params().keys()

grid_search = GridSearchCV(optimal_rfc, param_grid)
grid_search.fit(X_for_train, y_for_train)
print(grid_search.best_params_)

better_model = grid_search.best_estimator_
better_pred = better_model.predict(X_for_test)
better_accuracy = accuracy_score(y_for_test, better_pred)
print(better_accuracy)
```

We can see some linearity betweeen the "n_estimators" parameter, it will increase the probability up to some point (we could add 400, 500...) but it's time consuming.

### eXtreme Gradient Boosting hyper parameter tuning

```python
param_grid = {'max_depth': [3,10,20],
              'learning_rate':[0.3],
              'booster':['gbtree'],
              'n_estimators':[100,200],
                'n_jobs':[10],
             'eval_metric':['mlogloss']}

# all the numbers in param_grid are for an optimal parameter tuning 
# tested with many possibilities and provided the best amongest them 
#note: these params vary from problem to problem. These are some of best suitable for this problem 
   
optimal_xgbc = XGBClassifier()
param = optimal_xgbc.get_params().keys()

grid_search = GridSearchCV(optimal_xgbc, param_grid)
grid_search.fit(X_for_train, y_for_train)
print(grid_search.best_params_)

better_model = grid_search.best_estimator_
better_pred = better_model.predict(X_for_test)
better_accuracy = accuracy_score(y_for_test, better_pred)
print(better_accuracy)
```

### Linear Support Vector hyper parameter tuning

```python
param_grid = {'max_iter': [100, 1000],
               'random_state': [0,50,100], 
              'tol':[0.01, 0.1, 0.5 ], 
              'C':[0.5, 1, 1.5]}

# all the numbers in param_grid are for an optimal parameter tuning 
# tested with many possibilities and provided the best amongest them 
#note: these params vary from problem to problem. These are some of best suitable for this problem 
           
optimal_svc = svm.LinearSVC()
param = optimal_svc.get_params().keys()

grid_search = GridSearchCV(optimal_svc, param_grid)
grid_search.fit(X_for_train, y_for_train)
print(grid_search.best_params_)

better_model = grid_search.best_estimator_
better_pred = better_model.predict(X_for_test)
better_accuracy = accuracy_score(y_for_test, better_pred)
print(better_accuracy)
```
